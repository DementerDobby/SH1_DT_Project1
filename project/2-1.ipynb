```python
import pandas as pd

# '2_1.csv' 파일을 'euc-kr' 인코딩으로 읽어서 DataFrame으로 저장
df = pd.read_csv('2_1.csv', encoding='euc-kr')

# 데이터프레임 내용 출력
print(df)
```

```python
# 컬럼 이름만 추출
columns = df.columns.tolist()

# 컬럼 이름 출력
print(columns)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 준비
status_counts = df['진행상태'].value_counts()

# 파이 차트 생성
plt.figure(figsize=(10, 8))
plt.pie(status_counts, 
        labels=status_counts.index, 
        autopct='%1.1f%%', 
        colors=plt.cm.Set2(range(len(status_counts))), 
        startangle=140, 
        wedgeprops=dict(width=0.4))  # 원형 차트의 두께 조정

# 제목 추가
plt.title('진행상태별 비율', fontsize=16)

# 범례 추가
plt.legend(title='진행상태', bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=12)

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 예를 들어 데이터프레임을 생성합니다.
# df = pd.read_csv('your_data.csv')  # 실제 데이터로 교체하세요.

plt.figure(figsize=(14, 7))

# 막대 그래프
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='진행상태', palette='Set2')
plt.xlabel('진행상태')
plt.ylabel('항목 수')
plt.title('진행상태별 항목 수')

# 파이 차트
plt.subplot(1, 2, 2)
status_counts = df['진행상태'].value_counts()
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', colors=plt.cm.Set2(range(len(status_counts))), startangle=140)
plt.title('진행상태별 비율')

plt.tight_layout()
plt.show()
```
