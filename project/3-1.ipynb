```python
import matplotlib.pyplot as plt

# 설계기능별 D/P&BOM 내역 비율
design_function_bom = df.groupby('설계기능')['D/P&BOM 내역'].value_counts().unstack().fillna(0)

# 파이 차트로 시각화
plt.figure(figsize=(14, 7))
for idx, function in enumerate(design_function_bom.index):
    plt.subplot(1, len(design_function_bom.index), idx + 1)
    data = design_function_bom.loc[function]
    
    if not data.empty and data.sum() > 0:
        plt.pie(data, labels=data.index, autopct='%1.1f%%', colors=plt.cm.Set2(range(len(data))), startangle=140)
        plt.title(function)
    else:
        plt.text(0.5, 0.5, 'No Data', horizontalalignment='center', verticalalignment='center', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# 설계기능별 도면종류 집계
design_function_drawing_counts = df.groupby('설계기능')['도면종류'].value_counts().unstack().fillna(0)

# 스택형 막대그래프 생성
design_function_drawing_counts.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='Set2')
plt.xlabel('설계기능')
plt.ylabel('빈도수')
plt.title('설계기능별 도면종류 분포')
plt.legend(title='도면종류')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
